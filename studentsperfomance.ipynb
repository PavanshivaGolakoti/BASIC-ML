#**Visualization**
# **COUNT** **PLOT**

```python
plt.title('No. of Groups')
plot=sns.countplot(x="race/ethnicity",data=df)
```

From this visualization,we can observe that our dataset is not balanced as we don't have equal number of count of the groups. We can also observe that we have more number in group B  maintaining a huge differnce with the count in group 

# **SCATTER PLOT**

```python
fig, axs = plt.subplots(figsize = (10,10))
plot=sns.scatterplot(x='reading score',y='writing score',hue="test preparation course",data=df)
```

From the above visualization, we can tell that the students who have completed the preparaton for exam scored good marks and students who did not completed the preparation for exam got low marks but some students who did  not completed preparation also got good marks
# **STRIP PLOT**

```python
fig, axs = plt.subplots(figsize = (6,6))
plot=sns.stripplot(x="test preparation course",y="math score",hue="gender",data=df);
```

From the above visualization,we observed that the male students has higher marks  when compared to female student marks in the two courses of our dataset.

```python
fig, axs = plt.subplots(figsize = (8,6))
plot=sns.boxplot(x="test preparation course",y="math score",data=df)
```

Here,we can observe that both the completed and not completed preparation students have got almost equal marks but the  preparation completed students are little bit high in scoring
# **Correlation Between Variables**

```python
df.corr()
```

Further, we can visualise this correlation using a heatmap.
# **Heat Map**

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(df.corr(), cmap='Reds', annot = True);
```

# **HIST PLOT**

```python
df.hist(edgecolor='black',bins = 25, figsize= (12,6))
plt.show()
```

Here,we can visualize the distribution of scores of our dataset.
# **PAIR PLOT**

```python
sns.pairplot(data = df,hue="gender", height=3)
```

pairplot is showing the bivariate relation between each pair of features. From the pairplot, we'll see that the Gento species is separataed from the other two across all feature combinations

# **JOINT PLOT**

```python
sns.jointplot(x="reading score", y="writing score", data=df,height=5)
plt.show()
```

We can also use the seaborn library to make a similar plot. A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure.

# **DIST PLOT**

```python
sns.distplot(df['writing score'],bins=5)
```

From the above Dist plot, we can easily observe that most students got marks in range  from 60 to 80 and  only few  students got above 90
